```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
BASE_DIR = "data/BRCA"
```

```python
names_dict = {
    "methy":"1", "mirna":"2","mrna":"3"
}


names_dict_rev = {v: k for k, v in names_dict.items()}
names_dict, names_dict_rev
```

```python
for k, v in names_dict.items():
    featname_filename = f"{v}_featname.csv"
    te_filename = f"{v}_te.csv"
    tr_filename = f"{v}_tr.csv"
    label_test_filename = "labels_te.csv"
    label_train_filename = "labels_tr.csv"
    featname_path = os.path.join(BASE_DIR, featname_filename)
    te_path = os.path.join(BASE_DIR, te_filename)
    tr_path = os.path.join(BASE_DIR, tr_filename)
    lb_te_path = os.path.join(BASE_DIR, label_test_filename)
    lb_tr_path = os.path.join(BASE_DIR, label_train_filename)
    if not os.path.exists(featname_path) or not os.path.exists(te_path) or not os.path.exists(tr_path) or not os.path.exists(lb_te_path) or not os.path.exists(lb_tr_path):
        raise Exception(f"File {featname_path} does not exist")
    
    featname_df = pd.read_csv(featname_path, header=None)
    te_df = pd.read_csv(te_path, header=None)
    tr_df = pd.read_csv(tr_path, header=None)
    lb_tr_df  = pd.read_csv(lb_tr_path, header=None)
    lb_te_df  = pd.read_csv(lb_te_path, header=None)
    te_np = te_df.values
    tr_np = tr_df.values
    lb_tr_np = lb_tr_df.values
    lb_te_np = lb_te_df.values
    all_lb_np = np.concatenate([lb_tr_np, lb_te_np], axis=0)
    all_np = np.concatenate([tr_np, te_np], axis=0)
    print(f"For modalty: {k}")
    print(np.min(all_np), np.max(all_np), np.std(all_np), np.mean(all_np))
    final_data = np.concatenate([featname_df.values, all_np.T], axis=1).T
    final_data_df = pd.DataFrame(final_data)
    final_data_df.columns = final_data_df.iloc[0]
    final_data_df = final_data_df[1:]
    final_data_df['Label'] = all_lb_np
    train_labels = np.ones((lb_tr_np.shape[0], 1))
    test_labels = np.zeros((lb_te_np.shape[0], 1))
    all_labels = np.concatenate([train_labels, test_labels], axis=0)
    final_data_df['Split'] = all_labels
    final_data_df.to_csv(os.path.join(BASE_DIR,f"{k}.csv"))

```
