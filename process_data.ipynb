```python
import pandas as pd
import numpy as np
import os 
```

```python
train_data_path = '/home/dhakal/MoBI/data/LUAD/luad_train.csv'
test_data_path = '/home/dhakal/MoBI/data/LUAD/luad_test.csv'
features_path = '/home/dhakal/MoBI/data/LUAD/feature_name.csv'
```

```python
train_df  = pd.read_csv(train_data_path, header=None )
test_df = pd.read_csv(test_data_path, header=None)
feat_df = pd.read_csv(features_path, header=None)   

print(train_df.shape)
print(test_df.shape)
print(feat_df.shape)
```

```python
train_df['Split'] = '1'
test_df['Split'] = '0'
```

```python
feature_name = feat_df.iloc[:,0]
feature_name
```

```python
#concatenate the train and test data
df = pd.concat([train_df, test_df], axis=0)
df.shape
```

```python
df.head()
```

```python
df = df.rename(columns ={601: 'Label'})
```

```python
#rename the columns 1 to 600 to the feature names not the first column
df_new = df.rename(columns = dict(zip(df.columns[1:601], feature_name)))
df_new.drop(columns=[0], inplace=True)
# df = df.rename(columns = dict(zip(df.columns[0:600], feature_name)))
df_new.head()
```

```python
#make first 200 with label and split on the dataframe to mRNA df
mrna_df = df_new.iloc[:, np.r_[0:200, 600, 601]]

#save the mRNA df to csv
mrna_df.to_csv('/home/dhakal/MoBI/data/LUAD/mrna.csv', index=False)
mrna_df.head()
```
